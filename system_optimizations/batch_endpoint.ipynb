# 3_batch_endpoint

Evaluate the effective batch-size operating range of /predict_batch for bulk import workloads.

```python
from src.bench import load_config, build_batch_payload, run_fixed_concurrency
import pandas as pd
cfg = load_config()
```

```python
results = []
for bs in cfg['batch_size_candidates']:
    payload = build_batch_payload(cfg, bs)
    summary, _ = run_fixed_concurrency(
        cfg['base_url'], cfg['predict_batch_path'], payload,
        concurrency=1, total_requests=20, timeout=cfg['timeout_seconds']
    )
    results.append({'batch_size': bs, **summary})

results_df = pd.DataFrame(results)
results_df
```
