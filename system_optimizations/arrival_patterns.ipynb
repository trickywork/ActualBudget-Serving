# 2_arrival_patterns

After fixing the best worker configuration, compare constant and Poisson arrival patterns to evaluate how burstiness affects tail latency.

```python
from src.bench import load_config, build_single_payload, run_constant_rate, run_poisson_rate
import pandas as pd
cfg = load_config()
payload = build_single_payload(cfg)
```

```python
results = []
for rate in cfg['request_rate_candidates']:
    s1, _ = run_constant_rate(cfg['base_url'], cfg['predict_path'], payload, request_rate=rate, duration_s=20, timeout=cfg['timeout_seconds'])
    results.append({'pattern': 'constant', 'request_rate': rate, **s1})

    s2, _ = run_poisson_rate(cfg['base_url'], cfg['predict_path'], payload, request_rate=rate, duration_s=20, timeout=cfg['timeout_seconds'])
    results.append({'pattern': 'poisson', 'request_rate': rate, **s2})

results_df = pd.DataFrame(results)
results_df
```

```python
results_df.sort_values(['request_rate', 'pattern'])
```
