# 2_readiness_and_capacity
Goal: measure startup/readiness and burst capacity under representative load.


```python
import pandas as pd
from src.infra_bench import (
    load_config, build_single_payload, wait_until_ready,
    measure_first_request, run_request_rate_profile
)
cfg = load_config('config/infra_config.example.yaml')
payload = build_single_payload()
```

```python
# Run this immediately after starting or restarting the service.
ready_summary = wait_until_ready(cfg['base_url'], cfg['ready_path'], timeout_s=60)
first_req = measure_first_request(cfg['base_url'], cfg['predict_path'], payload, timeout=cfg['timeout_seconds'])
ready_summary, first_req
```

```python
capacity_results = []
for rate in cfg['request_rate_candidates']:
    for dist in ['constant', 'poisson']:
        summary, _ = run_request_rate_profile(
            cfg['base_url'], cfg['predict_path'], payload,
            request_rate=rate, duration_s=20,
            distribution=dist, timeout=cfg['timeout_seconds']
        )
        capacity_results.append(summary)
capacity_df = pd.DataFrame(capacity_results)
capacity_df
```
