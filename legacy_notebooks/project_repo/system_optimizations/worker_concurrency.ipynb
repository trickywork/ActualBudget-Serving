# 1_worker_concurrency

Fix the best model and identify a stable operating range of worker × concurrency for the online /predict endpoint.

```python
from src.bench import load_config, build_single_payload, run_fixed_concurrency
cfg = load_config()
payload = build_single_payload(cfg)
cfg
```

```python
import pandas as pd
results = []
for workers in cfg['worker_candidates']:
    print(f'\n=== 请先用 workers={workers} 启动服务，再运行下面单元 ===')
```

```python
results = []

for workers in cfg['worker_candidates']:
    for conc in cfg['concurrency_candidates']:
        summary, _ = run_fixed_concurrency(
            cfg['base_url'],
            cfg['predict_path'],
            payload,
            concurrency=conc,
            total_requests=80,
            timeout=cfg['timeout_seconds'],
            workers=workers,  
        )
        results.append({
            'workers': workers,
            'concurrency': conc,
            **summary
        })

results_df = pd.DataFrame(results).sort_values(
    by=['workers', 'concurrency']
).reset_index(drop=True)

results_df
```
